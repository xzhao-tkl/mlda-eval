```python
import pandas as pd

df = pd.read_csv("./ja_evaluation.annotated.50.csv")
```

```python
for col in ["faithfulness", 'fluency', 'diversity', 'factuality', 'inter-factuality']:
    print(f"{col}: {df[col].mean():.2f}")
```

```python
plau_sum = 0
for col in ['d1_plausibility', 'd2_plausibility', 'd3_plausibility']:
    plau_sum += df[col].mean()
print(f"plausibility: {plau_sum / 3:.2f}")


plau_sum = 0
for col in ['d1_incorrectness', 'd2_incorrectness', 'd3_incorrectness']:
    plau_sum += df[col].mean()
print(f"incorrectness: {plau_sum / 3:.2f}")
```
