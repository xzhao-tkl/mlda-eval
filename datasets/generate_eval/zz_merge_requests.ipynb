```python
sdf = df[~df['in-lang'].isna()]
sdf.shape
```

```python
for col in ['in-lang', 'cross-lang', 'faithfulness', 'factuality', 'fluency', 'equivalence', 'diversity']:
    print(f"{col}: {sdf[col].mean()}")
```

```python
plau_sum = 0
for col in ['d1_plausibility', 'p2_plausibility', 'p3_plausibility']:
    plau_sum += sdf[col].mean()
print(f"plausibility: {plau_sum / 3}")


plau_sum = 0
for col in ['p1_compatibility', 'p2_compatibility', 'p3_compatibility']:
    plau_sum += sdf[col].mean()
print(f"compatibility: {plau_sum / 3}")

plau_sum = 0
for col in ['p1_incorrectness', 'p2_incorrectness', 'p3_incorrectness']:
    plau_sum += sdf[col].mean()
print(f"incorrectness: {plau_sum / 3}")
```
