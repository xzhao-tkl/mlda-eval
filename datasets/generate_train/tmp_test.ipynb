```python
from utils import load_jsonl_iteratively

fn1 = "/data/xzhao/dataset/roman-pretrain/instructions/en_jstage/medical_native.jsonl"
fn2 = "/data/xzhao/dataset/roman-pretrain/noisy_instructions/en_jstage/wordnet.monolingual.32.jsonl"
```

```python
from tqdm import tqdm
from collections import Counter

c1_docs, c1_types = Counter(), Counter()
for item in tqdm(load_jsonl_iteratively(fn1)):
    c1_docs[item['docid']] += 1
    c1_types[item['type']] += 1

c2_docs, c2_types = Counter(), Counter()
for item in tqdm(load_jsonl_iteratively(fn2)):
    c2_docs[item['docid']] += 1
    c2_types[item['type']] += 1
```
