```python
import os
from tqdm import tqdm
from utils import load_jsonl_iteratively

lang = "en_jstage"
data_path = f"/data/xzhao/dataset/roman-pretrain/datasets/medical/{lang}/full.jsonl"
```

```python
## Find the document ids that involved in the training for sure and write them in to new file
line_cnts = 0
train_path = f"/data/xzhao/experiments/roman-pretrain/datasets/kg-datasets/{lang}-0.5/knowledge.jsonl"
for item in tqdm(load_jsonl_iteratively(train_path)):
    abstract = item['text']
    line_cnts += 1
    
print(f"line_cnts: {line_cnts}")

data_path = f"/data/xzhao/dataset/roman-pretrain/instructions/{lang}/medical_native.jsonl"
trained_instruction_ids = []
seen_elements = set()

for item in tqdm(load_jsonl_iteratively(data_path, request_num=line_cnts)):
    if item['docid'] not in seen_elements:
        trained_instruction_ids.append(item['docid'])
        seen_elements.add(item['docid'])
print("Number of documents involved in the training: ", len(trained_instruction_ids))

verified_id_filepath = f"/data/xzhao/experiments/roman-pretrain/datasets/kg-datasets/{lang}-0.5/verified_docids.jsonl"
with open(verified_id_filepath, 'w') as f:
    for docid in trained_instruction_ids:
        f.write(f"{docid}\n")
```

```python
assert lang == "en_jstage", "This code is specifically for the en_jstage dataset."
duplicated_doc_ids = []
for item in tqdm(load_jsonl_iteratively(data_path)):
    abstract = item['raw']['abstract']
    if abstract.startswith("Nutritional intake status and nutritional"):
        duplicated_doc_ids.append(item['docid'])
print("Number of duplicated documents: ", len(duplicated_doc_ids))


matched_doc_in_train_corpus = 0
line_cnts = 0
train_path = f"/data/xzhao/experiments/roman-pretrain/datasets/kg-datasets/{lang}-0.5/knowledge.jsonl"
for item in tqdm(load_jsonl_iteratively(train_path)):
    abstract = item['text']
    line_cnts += 1
    if "Nutritional intake status and nutritional" in abstract:
        matched_doc_in_train_corpus += 1

print(f"line_cnts: {line_cnts}")
print(f"matched_doc_in_train_corpus: {matched_doc_in_train_corpus}")

print("Duplicated documents in trained instruction IDs: ", set(duplicated_doc_ids).intersection(set(trained_instruction_ids)))
```
