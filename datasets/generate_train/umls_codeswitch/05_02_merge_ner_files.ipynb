```python
import os 
import json
from tqdm import tqdm
from utils import load_jsonl_iteratively

ner_tool = "bionlp"
ner_root = "/data/xzhao/dataset/roman-pretrain/datasets/medical/en_jstage/code-switch"
seen_docs = set()

target_filepath = os.path.join(ner_root, f"{ner_tool}_merged.jsonl")
assert not os.path.exists(target_filepath), f"File {target_filepath} already exists."

with open(target_filepath, "w", encoding="utf8") as outfile:
    for filename in tqdm(os.listdir(ner_root)):
        if not filename.endswith(".jsonl") or not filename.startswith(ner_tool):
            continue
        for item in load_jsonl_iteratively(os.path.join(ner_root, filename)):
            doc_id = item["docid"]
            if doc_id in seen_docs:
                continue
            seen_docs.add(doc_id)
            string = json.dumps(item, ensure_ascii=False)
            outfile.write(f"{string}\n")
```

```python
# Filter merged file of version 0 using the newly filtered data.json file that filtered out irrelevant documents
ner_tool = "scibert"

target_filepath = os.path.join(ner_root, f"{ner_tool}_merged.jsonl")
assert not os.path.exists(target_filepath), f"File {target_filepath} already exists."

raw_datapath = "/data/xzhao/dataset/roman-pretrain/datasets/medical/en_jstage/data.jsonl"
doc_ids = set([item["docid"] for item in load_jsonl_iteratively(raw_datapath)])
print("Total documents in raw data:", len(doc_ids))

version0_file = os.path.join(ner_root, f"{ner_tool}_merged.jsonl.0")
assert os.path.exists(version0_file), f"File {version0_file} does not exist."
with open(target_filepath, "w", encoding="utf8") as outfile:
    for item in tqdm(load_jsonl_iteratively(version0_file), desc="Filtering NER files"):
        doc_id = item["docid"]
        if doc_id not in doc_ids:
            continue
        string = json.dumps(item, ensure_ascii=False)
        outfile.write(f"{string}\n")
```
