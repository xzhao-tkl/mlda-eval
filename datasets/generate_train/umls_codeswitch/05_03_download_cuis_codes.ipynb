```python
import os
from collections import Counter
from utils import load_jsonl_iteratively
```

```python
code_switch_root = "/data/xzhao/dataset/roman-pretrain/datasets/medical/en_jstage/code-switch/"
mltl_codes_filepath = os.path.join(code_switch_root, "cui_all-bionlp.jsonl")

all_cuis = 0
lang_counter = Counter()
for item in load_jsonl_iteratively(mltl_codes_filepath):
    if 'translations' in item and item['translations'] != []:
        lang_counter.update(item['translations'].keys())
    all_cuis += 1
```

```python
import pycountry

for lang, cnt in lang_counter.most_common(100):
    try:
        lang_name = pycountry.languages.get(alpha_3=lang.lower()).name
    except AttributeError:
        lang_name = lang  # fallback to code if not found
    print(f"{lang}-{lang_name}: {cnt} ({cnt/all_cuis:.2%})")
```
