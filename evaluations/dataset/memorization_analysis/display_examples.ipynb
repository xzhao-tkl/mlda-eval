```python
import random
from utils import load_jsonl

mem_items = load_jsonl("/data/xzhao/experiments/med-eval/dataset/adaxeval/ja_knowledge_memorization.jsonl")
gen_items = load_jsonl("/data/xzhao/experiments/med-eval/dataset/adaxeval/ja_knowledge_generalization.jsonl")
items = list(zip(mem_items, gen_items))

random.seed(1234)
items = random.sample(items, k=10)

def decoprate(text):
    return "\\begin{CJK}{UTF8}{min} " + text + " \end{CJK}"

IDS = ['A', 'B', 'C', 'D']
for item, gen_item in items:
    # print(f"Context: {item['metadata']['sentence']}")
    # print(f"Cloze Query: {item['question']}")
    # print(f"Paraphrase: {gen_item['question']}")
    # options = '\n'.join(item['options'])
    # print(f"Options: \n{options}")
    # print(f"Answer: {item['answer_idx']}")
    # break
    context = decoprate(item['metadata']['sentence'])
    cloze_query = decoprate(item['question'])
    paraphrase = decoprate(gen_item['question'])

    options = [decoprate(f"{IDS[i]}. {option}") for i, option in enumerate(item['options'])]
    options = ' \\newline\n'.join(options)
    answer = decoprate(IDS[item['answer_idx']])
    string = f"{context} & {cloze_query} & {paraphrase} & {options} & {answer} \\\\\\hline"
    print(string)
    # print("-" * 80)
```
